```python
# Importando libs
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# Lendo arquivo csv e gerando daframe
df = pd.read_csv("data/data_tratado.csv")
```

```python
# Resumo estatístico do dataframe, com quartis, mediana, etc.
df['valor'].describe()
```

```python
# Excluir outlier
df = df.drop(df[df.valor == 3558].index)
```

```python
# Resumo estatístico do dataframe, com quartis, mediana, etc.
df['valor'].describe()
```

```python
# Excluir outlier
df = df.drop(df[df.valor == 2486].index)
```

```python
# Resumo estatístico do dataframe, com quartis, mediana, etc.
df['valor'].describe()
```

```python
# Exibindo dataframe
df
```

```python
# Grafico de dispersão entre idade x valor gasto
plt.scatter(df["valor"], df["IDADE_TRATADA"])
plt.title('Gráfico de Dispersão entre idade x valor gasto')
plt.show()
```

```python
# Grafico de dispersão entre função x valor gasto
plt.scatter(df["valor"], df["FUNCAO_TRATADA"])
plt.title('Gráfico de Dispersão entre funcao x valor gasto')
plt.show()
```
